```python
import cv2
import torch
from ultralytics import YOLO
```

```python
CONFIDENCE_THRESHOLD_LIMIT = 0.5
BOX_COLOUR = (0, 255, 0)
DEVICE = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
VIDEO_SOURCE = cv2.VideoCapture("C:\\Users\\yhd06\\OneDrive\\Desktop\\code\\iwaz\\yolo_pothole\\test_video\\KakaoTalk_20240808_174346328.mp4")
SAVE_FILE_PATH = "C:/Users/yhd06/OneDrive/Desktop/iwaz/pothole_data/output_frames"
```

```python
model = YOLO('yolov9s.pt')
model = YOLO('C:/Users/yhd06/OneDrive/Desktop/code/iwaz/yolo_pothole/runs/detect/train/weights/best.pt')
```

```python
fourcc = cv2.VideoWriter_fourcc(*'mp4v')  
output_file = "output.mp4" 
fps = 30

frame_width = int(VIDEO_SOURCE.get(cv2.CAP_PROP_FRAME_WIDTH))
frame_height = int(VIDEO_SOURCE.get(cv2.CAP_PROP_FRAME_HEIGHT))

out = cv2.VideoWriter(output_file, fourcc, fps, (frame_width, frame_height))
```

```python
while VIDEO_SOURCE.isOpened():
    success, frame = VIDEO_SOURCE.read()

    if success:
        results = model.track(frame, persist=True)

        annotated_frame = results[0].plot()

        cv2.imshow("YOLO Tracking", annotated_frame)

        if cv2.waitKey(1) & 0xFF == ord("q"):
            break
    else:
        break

VIDEO_SOURCE.release()
out.release()
cv2.destroyAllWindows()
```
