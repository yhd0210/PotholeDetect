```python
import torch
from ultralytics import YOLO
from PIL import Image
import matplotlib.pyplot as plt
import os
```

```python
CONFIDENCE_THRESHOLD_LIMIT = 0.5
BOX_COLOUR = (0, 255, 0)
DEVICE = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
IMG_SOURCE = './test_img/01.jpg'
#IMG_SOURCE = 'C:/Users/yhd06/OneDrive/Desktop/iwaz/pothole_data/output_pothole_dataset/val/images/AM_sunny_CI03_20210929_111835_194_1_aug_33.jpg'
#SAVE_FILE_PATH = "C:/Users/yhd06/OneDrive/Desktop/iwaz/pothole_data/output_frames"
```

```python
model = YOLO('yolov9s.pt')
model = YOLO('C:/Users/yhd06/OneDrive/Desktop/code/iwaz/yolo_pothole/runs/detect/train/weights/best.pt')
```

```python
test_images_folder = 'C:/Users/yhd06/OneDrive/Desktop/iwaz/pothole_data/output_pothole_dataset/test/images'

output_dir = './test_img_output' 
os.makedirs(output_dir, exist_ok=True)

for image_name in os.listdir(test_images_folder):
    image_path = os.path.join(test_images_folder, image_name)
    if os.path.isfile(image_path):
        #  
        img = Image.open(image_path)
        
        #   
        results = model(img)

        #    
        for i, result in enumerate(results):
            #   
            plt.imshow(result.plot())
            plt.axis('off')

            #   
            save_path = os.path.join(output_dir, f'{os.path.splitext(image_name)[0]}_result_{i}.jpg')
            plt.savefig(save_path, bbox_inches='tight', pad_inches=0)
            plt.close()

        print(f'Processed {image_name} and saved results to {output_dir}')
```
